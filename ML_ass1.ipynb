```python
for i in range(0,10):
  mapping_dict[i] = str(i)
```

here is the dictionary.

```python
for i in range(0,26):
  mapping_dict[10+i] = chr(65+i)
```

we will add more key value pairs to the above dictionary . Keys will begin from [10=i] where i will range from o to 26 which means we'll get keys from 10 t0 35.
values will be chr(65+i), these are the ASCII values - 65,66,67....of A,B,C.... respectively.

```python
mapping_dict
```

now the dictionary is here, but why did we create this dictionary?
the purpose is to generate random 36 numbers from 0 to 35 and take their values from the respective dictionary (0 to 9 then A to Z).

```python
random_number = random.randint(0,35)
print(mapping_dict[random_number])
```

random.randint(0, 35): This function randint() is a part of the random module. It generates a random integer between the specified range, which in this case is between 0 and 35. This means that the possible values for random_number will range from 0 to 35, and it could be any whole number within that range, including both 0 and 35.
And then we map this generated random number to the mapping_dict dictionary so that the value of the random number is printed from the dictionary.


```python
blank_string = str()
for i in range(0,10):
  coin_toss = random.randint(0,1)
  if coin_toss == 1:
     random_number = random.randint(0,35)
     blank_string = blank_string + mapping_dict[random_number]    #concatenate

     print(blank_string)
```

Created a blank string .
Generate a random coin toss with values o and 1. And if coin toss equals 1 then generate the same random number from 0 to 35. From mapping_dict, take the value of the generated random number(0 to 35) and concatenate it to the blank string created.
All this will be repeated 10 times by applying for loop in range(0,10) that is a maximun string of 10 characters can be generated randomly.
Print blank_string.

```python

reverse_mapping_dict = dict()

for k in mapping_dict.keys():
   reverse_mapping_dict[mapping_dict[k]]=k
```

Created an empty dictionary named reverse_mapping_dict. This dictionary will be the reverse of the mapping_dict such that values of the mapping_dict will be the keys of the reverse_mapping_dict and vice versa.



```python
reverse_mapping_dict
```

The reverse_mapping_dict will be such that now the keys will be string from 0 to 9 then A to Z and the values will be numbers from 0 to 35.

```python
highest_digit = max(blank_string)
highest_base = reverse_mapping_dict[highest_digit] + 1
```

Created a variable named highest_digit. It will store the max value of the blank string.(In Python, the max() function is used to find the maximum value among a collection of elements. It can take multiple arguments or an iterable (like a list, tuple, or set) and return the largest item in that collection.).


---
A new variable named highest_base is created. It will store the highest base value of the reverse_mapping_dict having highest_digit as the key.

```python
import random
```

```python
random_numbers = list()
for i in range(0,1000):
  blank_string = str()

  random_number = random.randint(0,35)
  blank_string = blank_string + mapping_dict[random_number]

  random_number = random.randint(0,35)
  blank_string = blank_string + mapping_dict[random_number]



  for i in range(0,8):
    coin_toss = random.randint(0,1)

    if coin_toss == 1:
      random_number = random.randint(0,35)
      blank_string = blank_string + mapping_dict[random_number]

    else:
      break
  random_numbers.append(blank_string)
```

A new empty list named random_numbers is created.
Now a loop is run 1000 times and two random numbers will be generated(as explained above) , mapping is done and two random values are stored in the blank_string.

Enters another loop for i in range(0,8): that runs 8 times. A coin toss is performed using random.randint. if the coin toss result is 1, it generates another random number between 0 and 35, accesses the corresponding value from mapping_dict, and appends it to the blank_string.
If coin toss result is 0, then loop breaks.


After generating the string (blank_string) with values obtained from mapping_dict, it appends the string to the random_numbers list.

```python
random_numbers
```

```python
base_frequency = dict()


for blank_string in random_numbers:
    highest_digit = max(blank_string)
    highest_base = reverse_mapping_dict[highest_digit] + 1

    if highest_base in base_frequency.keys():
      base_frequency[highest_base] += 1
    else:
      base_frequency[highest_base] = 1


    print("highest base of {} is {}".format(blank_string, highest_base))
```

A new empty dictionary named base_frequency is created. NOw we get highest base from the blank_string using the max() function (as already explained). Its just that now we have to do it for every single string present in random_numbers list.

Keys of base_frequency is the highest base from each string.
The values of base_frequency is 1 and if the highest_base repeats itself , 1 is added to the value. This loop continues for every blank_string in random_numbers.

Print highest_base of each blank_string.


```python
print(base_frequency)
```

base_frequency dictionary is printed with the highest_base (of the blank_string in random_numbers) as the keys and the frequency of each highest_base as the values.
